```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

df=pd.read_csv('/content/drive/MyDrive/Copy of sonar data.csv',header=None)
df.head()
```

```python
df.shape

df.describe()
df[60].value_counts()
df.groupby(60).mean()
```

```python
X=df.drop(columns=60,axis=1)
Y=df[60]
```

```python
x_train,x_text,y_train,y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
logis=LogisticRegression()
logis.fit(x_train,y_train)
```

```python
pred=logis.predict(x_train)
accuracy_score(pred,y_train)
```

```python
pred=logis.predict(x_text)
accuracy_score(pred,y_test)
```
